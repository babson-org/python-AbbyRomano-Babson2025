## 5. Dictionaries

```python
student = {
    'fname': 'Peggy',
    'lname': 'Lipton',
    'dob': '2006-08-30',
    'year': 'Senior',
    'attendance_rate': 0.92,
    'credits': 122
}
```

```python
print(student['year'])
```

```python
for key in student:
    print(key)
```

```python
for key in student.keys():
    print(key)
```

```python
for value in student.values():
    print(value)
```

```python
for key, value in student.items():
    print(f"{key} = {value}")
```

```python
print(student.get('gpa'))
```

```python
print(student.get('gpa', 'NEEDS TO STUDY'))
```

```python
student.setdefault('gpa', 0.0)
print(student)
```

## 6a. List of Lists

```python
board = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
```

```python
for row in board:
    print(row)
```

```python
for row in range(3):
    for col in range(3):
        print(str(board[row][col]) + ' ', end='')
    print()
```
